```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
from time import sleep
from selenium.webdriver.common.keys import Keys
from pandas.io.html import read_html
import pandas as pd
from selenium.webdriver.common.by import By


from bs4 import BeautifulSoup
```

```python
import warnings

url = 'https://www.agoda.com/city/seoul-kr.html'


chrome_options = Options()
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"]) # 셀레니움 로그 무시
warnings.filterwarnings("ignore", category=DeprecationWarning) # Deprecated warning 무시 
driver = webdriver.Chrome("/Users/jeongdain/Desktop/activity/crawling/chromedriver.exe") # webdriver 실행(크롬 창 생성)
driver.get(url) # url로 접속
time.sleep(2)
```

# 서울

```python
hotel_link=[]
hotel_links = driver.find_elements(By.CSS_SELECTOR, 'div.DatelessPropertyCard__Content > a')
for hotel in hotel_links:
    href = hotel.get_attribute('href')
    hotel_link.append(href)
print(hotel_link)
```

```python
   
review = []
hotel_name = []
star = []
total = []
grade = []


for hotel_url in hotel_link :   
        # 다시 request를 날려 주기
        # url = hotel_url + '#' + str(i)
        # # print(url)

        response = driver.get(hotel_url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        time.sleep(3)
        
        
        for i in range(1, 6): # 10page
            page = i
            # print(f'{i}페이지 크롤링 ing ========================== ')
            if i == 1: 

                post_reviews = driver.find_elements(By.CSS_SELECTOR,'div.Review-comment-body > p')
                for i in post_reviews:
                    post_review = i.text
                    review.append(post_review)

                post_scores = driver.find_elements(By.CSS_SELECTOR,'div.Review-comment-leftScore')
                for i in post_scores:
                    post_score = i.text
                    star.append(post_score)  
                    
                    name = driver.find_element(By.XPATH,'//*[@id="abouthotel-description"]/div[1]/p/b[1]/b').text
                    hotel_name.append(name)

                    # time.sleep(3)



            else : 
                
                driver.find_element(By.XPATH, '//*[@id="reviewSection"]/div[4]/div/span[2]/span['+str(i)+']').send_keys(Keys.ENTER)
                time.sleep(3)
                
                post_reviews = driver.find_elements(By.CSS_SELECTOR,'div.Review-comment-body > p')
                for i in post_reviews:
                    post_review = i.text
                    review.append(post_review)

                post_scores = driver.find_elements(By.CSS_SELECTOR,'div.Review-comment-leftScore')
                for i in post_scores:
                    post_score = i.text
                    star.append(post_score)  
                    
                    name = driver.find_element(By.XPATH,'//*[@id="abouthotel-description"]/div[1]/p/b[1]/b').text
                    hotel_name.append(name)
            
            
            
            
            print('%d번째 %s호텔 크롤링 ing ========================== '%(page,name))


        # print(html) 



```

```python
df = pd.DataFrame({
    'review': review,
    'score' : star,
    'name': hotel_name,
    # 'total' : total,
    # 'grade' : grade
    })
df
```

```python
df.to_csv('/Users/jeongdain/Desktop/activity/crawling/agoda_hotel_review_english_찐.csv',index=False)
```
